Add project in Google Drive (current logged-in account) as local library (need permission)

```python
import sys as Environment
import os as OS
import google.colab.drive as Drive_CoLab
import os.path as Pathing

PlaceHolder_CurrentFolder=F"."
Directory_Current=Pathing.abspath(PlaceHolder_CurrentFolder)
#
print(F"{Directory_Current=}")
print(F"{OS.getcwd()=}")

Directory_Base=F"/content"
Directory_Drive=F"{Directory_Base}/drive"
#
Folder_Drive_Google=F"MyDrive"
#Folder_Drive_Google=F"My Drive"
#
Directory_WorkBook=F"{Directory_Drive}/{Folder_Drive_Google}/Colab Notebooks"
Folder_AIS=F"Library_AIS"
Directory_AIS=F"{Directory_WorkBook}/{Folder_AIS}"
#
print(F"{Directory_Drive=}")
print(F"{Directory_WorkBook=}")
print(F"{Directory_AIS=}")

Drive_CoLab.mount\
(
    Directory_Drive,
    force_remount=True,
)
#
#Environment.path.append(Directory_Base_CoLab)
Environment.path.append(Directory_AIS)

OS.chdir(Directory_AIS)
# OS.chdir(Directory_WorkBook)
#
print(F"{OS.getcwd()=}")
```
